# Relational Plot = relplot.
The Seaborn relplot() function is used to create figure-level relational plots onto a Seaborn FacetGrid.


```python
sns.relplot(x='year',y='passengers',data=flights)
```

```python
# Number of Passengers Over Time by Month with scatter Plot -
sns.relplot(x='year', y='passengers', kind='scatter', hue='month', data=flights)
```

```python
# Number of Passengers Over Time by Month with Line Plot -
sns.relplot(x='year', y='passengers', kind='line', hue='month', data=flights)
```

Scatter Plot -
Scatter plot is one of the graphs that helps users to indicate each and every data value on the plot.

```python
sns.scatterplot(x='year',y='passengers', hue='month', data=flights, palette='bright')
```

Categorical Plot = Cat Plot.

This function provides access to several axes-level functions that show the relationship between a numerical and one or more categorical variables using one of several visual representations.

```python
# Passengers Count Over Months
sns.catplot(x='month',y='passengers',data=flights, color='green')
```

```python
# Passengers Count Over Months
sns.catplot(x='month',y='passengers',data=flights, color='green',kind='swarm',hue='month')
```

Bar Plot - A barplot is a type of plot that displays the numerical values for different categorical variables.

```python
sns.barplot(x="month", y="passengers", data=flights, ci=None, palette="Greens_d")
```

Count Plot - A countplot basically counts the categories and returns a count of their occurrences.

```python
# Number of flights for each Month
sns.countplot(x="year", data=flights, palette="bright", )
```

Distplot - This function provides access to several approaches for visualizing the univariate or bivariate distribution of data, including subsets of data defined by semantic mapping and faceting across multiple subplots.

```python
sns.distplot(flights['passengers'],color='green')
```

```python
sns.distplot(flights['passengers'],kde=True,hist=False,color='green')
```

Box Plot - A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
sns.boxplot(x="month", y="passengers", data=flights, color="green")
```

```python
# Box Plot with Help of Cat Plot using Kind function.
sns.catplot(data=flights,x='month',y='passengers',kind='box', color='yellow')
```

```python
sns.violinplot(data=flights,x='month',y='passengers',color='purple')
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(flights['passengers'],bins=8, color='green')
plt.show()
```
